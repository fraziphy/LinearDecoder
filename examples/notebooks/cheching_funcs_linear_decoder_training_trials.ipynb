```python
spikes_trials_all = []
for trial in range(1,21):
    spike_1 = generate_data.dummy_spiketimes(n_neurons, duration, dt, avg_rate)
    spikes_trials_all.append(spike_1)
```

```python
n_neurons = 10000
duration=200
dt=0.1
n_steps = int(duration / dt)
time = np.arange(n_steps) * dt

# Generate base sine wave with a period of 400 ms
frequency = 2.5  # 2.5 Hz for 400 ms period
base_sine = 2 * np.sin(2 * np.pi * frequency * time / 1000)

# Keep only the upper half of the sine wave
signal = np.maximum(base_sine, 0)
```

```python
# Initialize
decoder = funcs.LinearDecoder(dt=0.1, tau=10, lambda_reg=1e-3, rng=np.random.default_rng(2))

# Preprocess data
filtered_spikes = decoder.preprocess_data(spikes_trials_all, n_neurons, duration)

# Perform stratified cross-validation
train_errors, test_errors, all_weights = decoder.stratified_cv(filtered_spikes, signal)

# Print results
print("Average Train RMSE:", train_errors.mean(axis=0))
print("Average Test RMSE:", test_errors.mean(axis=0))
```
